```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
numerical_col = df.select_dtypes(include=["number"]).columns
for col in numerical_col:
  if col == 'Precipitation Corrected (mm/day)':
    continue
  else:
    skewness = df[col].skew()
    print(skewness)
    if skewness == 0:
      print(f"Finding outliers in {col} using Z-score method")
      mean = df[col].mean()
      std_dev = pd.DataFrame[col].std()
      # Identify outliers (Z-scores < -3 or > 3)
      df = df[((df[col] - mean) / std_dev).abs() < 3]
      # print(f"Outliers in {col}:\n",outliers)
      # print("-------"*50)
    else:
      print(f"Finding outliers in {col} using IQR method")
      Q1 = df[col].quantile(0.25, interpolation='nearest')
      Q2 = df[col].quantile(0.50, interpolation='midpoint')
      Q3 = df[col].quantile(0.75, interpolation='nearest')
      IQR = Q3 - Q1
      print("IQR: ",IQR)
      threshold=1.5*IQR
      lower_bound=Q1-threshold
      upper_bound=Q3+threshold
      print("Lower Bound:",lower_bound, "Upper Bound:" ,upper_bound)
      #storing the entries that are not outliers
      df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
print(f"After removing outliers:\n",df)
df.shape

```

```python
#datetime required columns with names year,month,day so renaming YEAR,MO,DY to that
df['date'] = pd.to_datetime(df[['YEAR', 'MO', 'DY']].rename(columns={'YEAR': 'year', 'MO': 'month', 'DY': 'day'}))
df = df.drop(columns=['YEAR', 'MO', 'DY'])
# Display the DataFrame
print(df)
```
